```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import ActionChains
import time
```

```python
class InstagramBot():
    
    ##Constructor
    def __init__(self, username, password):
        self.driver = webdriver.Firefox(executable_path = 'C:\webdriver\geckodriver.exe')
        self.username = username
        self.password = password
        self.driver.implicitly_wait(10)
    
    
    ##SignIn
    def signIn(self):
        self.driver.get('https://www.instagram.com/?hl=en')
        emailInput = self.driver.find_element_by_name('username')
        passwordInput = self.driver.find_element_by_name('password')
        emailInput.send_keys(self.username)
        passwordInput.send_keys(self.password)
        passwordInput.send_keys(Keys.ENTER)
        self.driver.find_element_by_class_name('HoLwm').click()
        time.sleep(2)
    
    
    ##Name of all Food Handles without HashTag
    def FoodHandlesName(self):
        time.sleep(3)
        self.driver.find_element_by_class_name('Di7vw').click()
        text_box = self.driver.find_element_by_class_name('x3qfX')
        text_box.send_keys('food')
        drp_dwn = self.driver.find_elements_by_class_name('Ap253')
        for i in drp_dwn:
            if(i.text[0] == '#'):
                continue
            print(i.text)
        self.driver.find_element_by_css_selector('div.aIYm8.coreSpriteSearchClear').click()
    
    
    ##So delhi Profile
    def SoDelhi(self):
        time.sleep(3)
        self.driver.find_element_by_class_name('Di7vw').click()
        text_box = self.driver.find_element_by_class_name('x3qfX')
        text_box.send_keys('So Delhi')
        b = self.driver.find_elements_by_css_selector('span.Ap253')
        for i in b:
            if(i.text == 'sodelhi'):
                i.click()
                break
    
    
    
    ##Follow the profile (if we want to follow So delhi if open so delhi profile and run this function)
    def Follow(self):
        time.sleep(3)
        followButton = self.driver.find_element_by_css_selector('._5f5mN    ')
        if (followButton.text == 'Follow'):
            followButton.click()
            time.sleep(2)
        else:
            print("You are already following this user")
    
    
    ####UnFollow the profile (if we want to follow So delhi if open so delhi profile and run this function)
    def UnFollow(self):
        time.sleep(3)
        UNfollowButton = self.driver.find_element_by_css_selector('._5f5mN    ')
        #print(followButton.text)
        if (UNfollowButton.text == ''):
            UNfollowButton.click()
            self.driver.find_element_by_css_selector('.-Cab_   ').click()
            time.sleep(2)
        else:
            print("You are already unfollowing this user")
    
    
    ##Like the 30 post of dilsefoodie
    def Like(self):
        time.sleep(3)
        self.driver.get('https://www.instagram.com/dilsefoodie/?hl=en')
        a = self.driver.find_element_by_class_name('_bz0w')
        a.click()
        for i in range(1,31):
            Like = self.driver.find_element_by_css_selector('section.Slqrh span.fr66n button svg')
            if(Like.get_attribute('aria-label') == 'Like'):
                Like.click()
            else:
                print('You are already Like this post ',i)
            a = self.driver.find_element_by_class_name('coreSpriteRightPaginationArrow')
            a.click()
            time.sleep(4)
        self.driver.find_element_by_class_name('TxciK').click()
    
    
    ##UnLike the 30 post of dilsefoodie
    def UnLike(self):
        time.sleep(3)
        self.driver.get('https://www.instagram.com/dilsefoodie/?hl=en')
        a = self.driver.find_element_by_class_name('_bz0w')
        a.click()
        for i in range(1,31):
            Unlike = self.driver.find_element_by_css_selector('section.Slqrh span.fr66n button svg')
            #print(Like.get_attribute('aria-label'))
            if(Unlike.get_attribute('aria-label') == 'Unlike'):
                Unlike.click()
            else:
                print('You are already Unlike this post ',i)
            a = self.driver.find_element_by_class_name('coreSpriteRightPaginationArrow')
            a.click()
            time.sleep(4)
        self.driver.find_element_by_class_name('TxciK').click()
    
    
    ##Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.
    def foodtalkindia_SoDelhi500(self):
        
        ##foodtalkindia
        print("foodtalkindia\n")
        time.sleep(3)
        self.driver.find_element_by_class_name('Di7vw').click()
        text_box = self.driver.find_element_by_class_name('x3qfX')
        text_box.send_keys('foodtalkindia')
        b = self.driver.find_elements_by_css_selector('span.Ap253')
        for i in b:
            if(i.text == 'foodtalkindia'):
                i.click()
            break
        time.sleep(2)
        c = self.driver.find_elements_by_css_selector('a.-nal3')
        for i in c:
            if(i.text.split()[1] == 'followers'):
                i.click()
                break
        followersList = self.driver.find_element_by_css_selector('div[role=\'dialog\'] ul')
        numberOfFollowersInList = len(followersList.find_elements_by_css_selector('li'))  
        followersList.click()
        actionChain = ActionChains(self.driver)
        max = 500
        while (numberOfFollowersInList < max):
            actionChain.key_down(Keys.SPACE).key_up(Keys.SPACE).perform()
            numberOfFollowersInList = len(followersList.find_elements_by_css_selector('li'))        
        followers = []
        for user in followersList.find_elements_by_css_selector('li'):
            userLink = user.find_element_by_css_selector('a._0imsa').get_attribute('title')
            print(userLink)
            followers.append(userLink)
            if (len(followers) == max):
                break
        print(followers)

        ##Close the popup
        self.driver.find_element_by_css_selector('svg[aria-label = "Close"]').click()
        
        
        ##So Delhi
        print("\n\nSo Delhi\n\n")
        time.sleep(3)
        self.driver.find_element_by_class_name('Di7vw').click()
        text_box = self.driver.find_element_by_class_name('x3qfX')
        text_box.send_keys('So Delhi')
        b = self.driver.find_elements_by_css_selector('span.Ap253')
        for i in b:
            if(i.text == 'sodelhi'):
                i.click()
            break
        time.sleep(2)
        c = self.driver.find_elements_by_css_selector('a.-nal3')
        for i in c:
            if(i.text.split()[1] == 'followers'):
                i.click()
                break
        followersList = self.driver.find_element_by_css_selector('div[role=\'dialog\'] ul')
        numberOfFollowersInList = len(followersList.find_elements_by_css_selector('li'))  
        followersList.click()
        actionChain = ActionChains(self.driver)
        max = 500
        while (numberOfFollowersInList < max):
            actionChain.key_down(Keys.SPACE).key_up(Keys.SPACE).perform()
            numberOfFollowersInList = len(followersList.find_elements_by_css_selector('li'))        
        followers = []
        for user in followersList.find_elements_by_css_selector('li'):
            userLink = user.find_element_by_css_selector('a._0imsa').get_attribute('title')
            print(userLink)
            followers.append(userLink)
            if (len(followers) == max):
                break
        print(followers)

        ##Close the popup
        self.driver.find_element_by_css_selector('svg[aria-label = "Close"]').click()
        
    
                                          
    ##Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.                                      
    def foodtalkindiafollowing(self):
        time.sleep(3)
        followers = []
        self.driver.get('https://www.instagram.com/foodtalkindia/?hl=en')
        a = []
        followersLists = self.driver.find_elements_by_css_selector('span.tc8A9 span._32eiM')
        for i in followersLists:
            a.append(i.text)
        self.driver.get('https://www.instagram.com/nikhildadhich1997/')
        time.sleep(5)
        c = self.driver.find_elements_by_css_selector('a.-nal3')
        for i in c:
            if(i.text.split()[1] == 'followers'):
                i.click()
                break
        time.sleep(4)
        followersList = self.driver.find_element_by_css_selector('div[role=\'dialog\'] ul')
        numberOfFollowersInList = len(followersList.find_elements_by_css_selector('li'))  
        max = int(self.driver.find_element_by_css_selector('a.-nal3 span.g47SY').get_attribute('title'))
        followersList.click()
        actionChain = ActionChains(self.driver)
        while (numberOfFollowersInList < max):
            actionChain.key_down(Keys.SPACE).key_up(Keys.SPACE).perform()
            numberOfFollowersInList = len(followersList.find_elements_by_css_selector('li'))        
        for user in followersList.find_elements_by_css_selector('li'):
            userLink = user.find_element_by_css_selector('a._0imsa').get_attribute('title')
            followers.append(userLink)
            if (len(followers) == max):
                break

        ##Close the popup
        self.driver.find_element_by_css_selector('svg[aria-label = "Close"]').click()
        for i in followersLists:
            for j in followers:
                if(i == j):
                    print(i)
                    break
    
                                          
                                          
    ##Coding Ninjas Story
    def CodingNinjasStory(self):
        time.sleep(3)
        #https://www.instagram.com/coding.ninjas/
        self.driver.get('https://www.instagram.com/coding.ninjas/?hl=en')
        CheckStory = self.driver.find_element_by_css_selector('header.vtbgv>div.XjzKX>div.RR-M-')
        a=CheckStory.get_attribute('class')
        if('h5uC0' in a):
            Story = self.driver.find_element_by_css_selector('header.vtbgv div.XjzKX div.h5uC0.RR-M- canvas.CfWVH')
            if(Story.get_attribute('width') == '249'):
                print('You have already seen the story')
            elif(Story.get_attribute('width') == '252'):
                print('You have not seen the story')
                StoryClick = self.driver.find_element_by_css_selector('header.vtbgv div.XjzKX div.h5uC0.RR-M- span._2dbep img._6q-tv')
                StoryClick.click()
        else:
            print('The user has no story')
```
